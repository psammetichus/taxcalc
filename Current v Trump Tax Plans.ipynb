# Tax Plans

This is a demonstration of the effects of the 2016 projected US tax brackets and those of the proposed Trump tax plan. Both are calculated for married couples filing jointly taking a standard deduction. No FICA or Medicare tax has been calculated and no investment income is reported.

```python
import numpy as np
```

```python
import tax
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
incomes = np.arange(10000., 3000000., 2000.)
```

## Tax rates as a function of income

```python
rates_a = tax.calctax(tax.us2016, incomes)/ incomes
```

```python
rates_b = tax.calctax(tax.trump, incomes)/ incomes
```

### US 2016 Rates

```python
plt.plot(incomes, rates_a)
```

### Trump rates

```python
plt.plot(incomes, rates_b)
```

## Absolute and relative differences (Trump - Standard) 

```python
plt.plot(incomes, tax.calctax(tax.trump,incomes) - tax.calctax(tax.us2016, incomes))
```

```python
plt.plot(incomes, (tax.calctax(tax.trump,incomes) - tax.calctax(tax.us2016, incomes))/incomes)
```
